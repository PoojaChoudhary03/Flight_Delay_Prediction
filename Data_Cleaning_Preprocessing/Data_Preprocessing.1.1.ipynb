```python
#import the libraries
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv('flight_data.csv')
```

```python
df.head()
```

```python
#Binary Classification
status = []

for value in df['ARR_DELAY']:
    if value < 0:
        status.append(0)
    else:
        status.append(1)
df['ARR_DELAY'] = status
df.head(2)    
```

```python
df = df.drop(["Unnamed: 0"], axis = 1)
```
